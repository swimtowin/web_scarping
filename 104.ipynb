```python
from selenium import webdriver 
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import json
import os 

options = webdriver.ChromeOptions()
options.add_argument("--incognito")
options.add_argument("--disable-popup-blocking")

driver = webdriver.Chrome(options = options)

listData = []

def visit():
    driver.get("https://www.104.com.tw/jobs/main/");

def search():
    txtInput = driver.find_element(By.CSS_SELECTOR, "#ikeyword")
    txtInput.send_keys("python")
    
    sleep(2)
                                  
def filterFunc():
    try:

        driver.find_element(
            By.CSS_SELECTOR,
            "#icity").click()
        
        
        sleep(2)
        
        driver.find_element(
            By.CSS_SELECTOR,
            "input[value='6001001000']").click()
        
        driver.find_element(
            By.CSS_SELECTOR,
            "input[value='6001002000']").click()


        WebDriverWait(driver,10).until(
        EC.presence_of_element_located(
                (By.CSS_SELECTOR, "button.category-picker-btn-primary")
            )        
        )  
            
            
        driver.find_element(
            By.CSS_SELECTOR,
            "button.category-picker-btn-primary").click()
        
        driver.find_element(
            By.CSS_SELECTOR,
            ".btn.btn-primary.js-formCheck").click()
        
    
        
        driver.find_element(
            By.CSS_SELECTOR,
            "div.b-float-left>ul#js-job-tab>li[data-value='1']").click()
       
        

    except TimeoutException:
        print("等待逾時，即將關閉瀏覽器...")
        sleep(3)
        driver.quit()
        
        
def scroll():
    innerHeightOfWindow = 0
    totalOffset = 0
    while totalOffset <= innerHeightOfWindow:
        totalOffset += 800;
        
        js_scroll = '''(
            function (){{
                window.scrollTo({{
                    top:{},
                    behavior: 'smooth'
                }});
            }})();'''.format(totalOffset)
    
        driver.execute_script(js_scroll)
        sleep(1)
        innerHeightOfWindow = driver.execute_script(
            'return window.document.documentElement.scrollHeight;'
        );

        print("innerHeightOfWindow: {}, totalOffset: {}".format(innerHeightOfWindow, totalOffset))
        if totalOffset >= 2000:
            break

def parse():
    article_info = driver.find_elements(
        By.CSS_SELECTOR,
        'article.b-block--top-bord.job-list-item.b-clearfix.js-job-item')
    for element in article_info:
        
        print("="*30)
        
        print("職缺名稱:")
        
        jobName = element.find_element(By.CSS_SELECTOR,"a.js-job-link")                         
        jobNametTitle = jobName.text
        print(jobNametTitle)
        
        print("職缺連結:")
        
        jobLink = jobName.get_attribute('href')
        print(jobLink)
        
        print("公司名稱:")
        
        companyName = element.find_element(By.CSS_SELECTOR,"a[title*='公司名']")
        companyNameTitle = companyName.text 
        print(companyNameTitle)
        
        print("公司連結:")
        
        companyLink = companyName.get_attribute('href')
        print(companyLink)
        
        print("地區:")
        
        location = element.find_element(By.CSS_SELECTOR,"ul.b-list-inline.b-clearfix.job-list-intro.b-content>li")
        locationTitle = location.text 
        print(locationTitle)
        
        listData.append({
            "職缺名稱:":jobNametTitle,
            "職缺連結:":jobLink,
            "公司名稱:":companyNameTitle,
            "公司連結:":companyLink,
            "地區:":locationTitle
        })
        
        
        
def saveJson():
    fp = open("104.json", "w", encoding='utf-8')
    fp.write( json.dumps(listData, ensure_ascii=False) )
    fp.close()
    
    
def close():
    driver.quit()
    
# def download():
    # 開啟 json 檔案
#     fp = open("104.json", "r", encoding='utf-8')
    
    #取得 json 字串
#     strJson = fp.read()
    
    # 關閉檔案
#     fp.close()
    
    # 將 json 轉成 list (裡面是 dict 集合)
    
    

       
        
        
if __name__ == '__main__':
    visit()
    search()
    filterFunc()
    scroll()
    parse()
    saveJson()
    close()
#     download()
```
